```python
import os

# 라벨링 파일이 있는 디렉터리 경로
directory = '파김치/train/labels'

# 0000 ~ 0999의 숫자 범위를 저장
expected_files = {f'{i:04d}' for i in range(1000)}

# 실제로 존재하는 파일의 숫자 부분을 저장
existing_files = set()

# 디렉터리 내 파일 순회
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        # 파일명에서 숫자 부분 추출 (예: Img_052_0000.txt -> 0000)
        number_part = filename.split('_')[-1].replace('.txt', '')
        existing_files.add(number_part)

# 비어 있는 파일 번호 찾기
missing_files = expected_files - existing_files

# 결과 출력
if missing_files:
    print("비어있는 파일 번호:")
    for missing in sorted(missing_files):
        print(missing)
else:
    print("비어 있는 파일 번호가 없습니다.")
```

```python
import os

# 라벨링 파일이 있는 디렉터리 경로
directory = '숙주나물/labels'

# 비어 있는 파일명을 저장할 리스트
empty_files = []

# 디렉터리 내 파일 순회
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        
        # 파일이 비어 있는지 확인
        if os.path.getsize(file_path) == 0:
            empty_files.append(filename)

# 결과 출력
if empty_files:
    print("비어 있는 파일:")
    for empty_file in empty_files:
        print(empty_file)
else:
    print("비어 있는 파일이 없습니다.")
```

```python
import os
# 라벨링 파일이 있는 디렉터리 경로
directory = '숙주나물/images'
# 파일 순회
for filename in os.listdir(directory):
    if filename.endswith('.jpg.jpg'):
        old_file_path = os.path.join(directory, filename)
        # 마지막 .jpg 제거 (뒤에 .jpg가 중복된 경우)
        new_filename = filename[:-4]  # 마지막 ‘.jpg’ 제거
        new_file_path = os.path.join(directory, new_filename)
        # 중복 파일이 있는 경우 덮어쓰기
        if os.path.exists(new_file_path):
            print(f"Removing existing file: {new_file_path}")
            os.remove(new_file_path)
        # 파일 이름 변경
        os.rename(old_file_path, new_file_path)
        print(f"Renamed {old_file_path} to {new_file_path}")
```
