이미지

```python
import os

# 이미지 파일이 있는 디렉터리 경로
images_directory = '깍두기/train/images'

# 변경하고자 하는 번호
old_number = '030'
new_number = '020'

# 이미지 파일 이름 변경 함수 (중간 번호만 변경)
def rename_image_files(directory, old_number, new_number):
    for filename in os.listdir(directory):
        # 파일 이름 형식이 'Img_040_0000.jpg'처럼 되어있다고 가정하고 중간 번호만 변경
        parts = filename.split('_')
        if len(parts) > 2 and parts[1] == old_number:
            parts[1] = new_number  # 중간 번호 변경
            new_filename = '_'.join(parts)
            old_file_path = os.path.join(directory, filename)
            new_file_path = os.path.join(directory, new_filename)
            
            # 중복 파일이 있는 경우 덮어쓰기
            if os.path.exists(new_file_path):
                os.remove(new_file_path)
            
            # 파일 이름 변경
            os.rename(old_file_path, new_file_path)
            print(f"{old_file_path} -> {new_file_path}")

# 이미지 파일 이름 변경
rename_image_files(images_directory, old_number, new_number)
print("이미지 파일 이름 변경이 완료되었습니다.")
```

라벨링

```python
import os

# 라벨링 파일이 있는 디렉터리 경로
labels_directory = '깍두기/train/labels'

# 변경하고자 하는 번호
old_number = '030'
new_number = '020'

# 라벨링 파일 이름 및 내용 변경 함수 (중간 번호만 변경)
def rename_label_files_and_modify_id(directory, old_number, new_number):
    for filename in os.listdir(directory):
        parts = filename.split('_')
        if len(parts) > 2 and parts[1] == old_number:
            parts[1] = new_number  # 중간 번호 변경
            new_filename = '_'.join(parts)
            old_file_path = os.path.join(directory, filename)
            new_file_path = os.path.join(directory, new_filename)
            
            # 중복 파일이 있는 경우 덮어쓰기
            if os.path.exists(new_file_path):
                os.remove(new_file_path)
            
            # 파일 이름 변경
            os.rename(old_file_path, new_file_path)
            print(f"{old_file_path} -> {new_file_path}")
            
            # 라벨링 파일 내용 변경 (첫 번째 값인 ID를 new_number로 변경)
            with open(new_file_path, 'r') as file:
                lines = file.readlines()
            
            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) > 0:
                    parts[0] = new_number  # 첫 번째 값(ID)을 new_number로 변경
                    new_lines.append(' '.join(parts) + '\n')
            
            # 변경된 내용을 파일에 다시 저장
            with open(new_file_path, 'w') as file:
                file.writelines(new_lines)

# 라벨링 파일 이름 및 ID 변경
rename_label_files_and_modify_id(labels_directory, old_number, new_number)
print("라벨링 파일 이름 및 ID 변경이 완료되었습니다.")
```

```python
import os

# 라벨링 파일이 있는 디렉터리 경로
labels_directory = '깍두기/train/labels'

# 라벨링 파일 내용에서 ID의 맨 앞 '0'을 삭제하는 함수
def remove_leading_zero_in_id(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)

            # 라벨 파일을 읽어서 수정
            with open(file_path, 'r') as file:
                lines = file.readlines()

            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) > 0:
                    # ID 값이 세 자리 숫자라면 맨 앞 '0'을 삭제
                    if parts[0].startswith('0') and len(parts[0]) == 3:
                        parts[0] = parts[0].lstrip('0')  # 맨 앞의 '0' 삭제
                    new_lines.append(' '.join(parts) + '\n')

            # 수정된 내용을 파일에 다시 저장
            with open(file_path, 'w') as file:
                file.writelines(new_lines)

            print(f"Updated ID in {filename}")

# 라벨링 파일 내용에서 ID 수정
remove_leading_zero_in_id(labels_directory)
print("ID의 맨 앞 '0' 삭제 완료")
```
