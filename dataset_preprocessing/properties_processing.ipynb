```python
import os

def split_properties_to_txt(kfood_path, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    
    for root, dirs, files in os.walk(kfood_path):
        for file in files:
            if file.endswith(".properties"):
                properties_file = os.path.join(root, file)
                with open(properties_file, 'r') as f:
                    lines = f.readlines()
                    
                for line in lines:
                    try:
                        img_name, bbox_str = line.strip().split('=')
                        bbox = list(map(int, bbox_str.split(',')))
                        if len(bbox) != 4:
                            print(f"잘못된 좌표값: {bbox_str} in {img_name}")
                            continue  # 좌표값이 올바르지 않은 경우 건너뛰기
                        
                        # 텍스트 파일로 저장
                        txt_file_path = os.path.join(output_dir, img_name + ".txt")
                        with open(txt_file_path, 'w') as txt_file:
                            txt_file.write(line)
                        
                    except Exception as e:
                        print(f"에러 발생: {e} in {line}")
                        continue

# 경로에 맞게 실행
split_properties_to_txt('kfood', 'output_txts')
```
